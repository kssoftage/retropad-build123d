```python
import trimesh
import numpy as np

files = {
    "bottom": "RetroPad - Bottom Shell.stl",
    "top":    "RetroPad - Top Shell.stl",
    "button": "RetroPad - Button.stl",
    "dpad":   "RetroPad - D-Pad.stl",
}

for name, f in files.items():
    m = trimesh.load(f)
    verts = m.vertices
    print(f"\n=== {name} ===")

    # Unique X values (rounded to 2dp) - reveals wall positions, hole centers
    ux = np.unique(np.round(verts[:,0], 2))
    print(f"  Unique X: {ux}")

    uy = np.unique(np.round(verts[:,1], 2))
    print(f"  Unique Y: {uy}")

    uz = np.unique(np.round(verts[:,2], 2))
    print(f"  Unique Z: {uz}")

    # For top shell - find circle centers (button holes, dpad)
    if name in ["top", "bottom"]:
        # Vertices at top Z level
        top_z = verts[:,2].max()
        top_verts = verts[np.abs(verts[:,2] - top_z) < 0.1]
        print(f"  Verts at top Z ({top_z:.2f}): X range {top_verts[:,0].min():.2f}→{top_verts[:,0].max():.2f}, Y range {top_verts[:,1].min():.2f}→{top_verts[:,1].max():.2f}")

        # Vertices at each Z level
        for z in uz:
            zv = verts[np.abs(verts[:,2] - z) < 0.05]
            print(f"  Z={z:.2f}: {len(zv)} verts, X=[{zv[:,0].min():.2f},{zv[:,0].max():.2f}], Y=[{zv[:,1].min():.2f},{zv[:,1].max():.2f}]")
```

```python
"""
RetroPad - Bottom Shell  (EXACT dimensions from STL analysis)
=============================================================
STL bounding box : 135.0 x 53.0 x 19.424 mm
STL volume       : 27602.85 mm³

Assembly coordinate system (preserved from original STL):
  X : -67.5  →  +67.5   (135 mm total)
  Y : -27.5  →  +25.5   (53 mm total, geometric centre at Y = -1.0)
  Z :  -6.0  → +13.42   (19.42 mm total height)

Exact Z levels found in mesh:
  -6.00  bottom of recessed foot
  -3.00  main bottom face
  -1.40  top of rear cable-notch slot
   0.00  inner floor
   3.60  snap / parting ring  (full outer XY extent)
  13.42  top rim face
"""

from build123d import *

# ── Exact coordinates (all from vertex analysis) ─────────────
# Outer shell extents
OX  =  67.5          # outer X half-span
YN  = -27.5          # outer Y negative limit
YP  =  25.5          # outer Y positive limit
YC  = (YN + YP) / 2  # = -1.0  (geometric centre in Y)

Z_BOTTOM   = -3.0    # main bottom face
Z_TOP      = 13.42   # top rim face
Z_SNAP     =  3.6    # snap ring Z (full outer XY)
Z_FLOOR    =  0.0    # inner cavity floor
Z_FOOT     = -6.0    # recessed foot bottom

# Inner cavity extents (at Z_FLOOR)
FX  =  55.0          # inner floor X half-span
FYN = -16.5          # inner floor Y negative
FYP =  15.0          # inner floor Y positive
FYC = (FYN + FYP) / 2

# Recessed foot extents
FTOX =  64.5
FTYN = -24.5
FTYP =  22.5
FTYC = (FTYN + FTYP) / 2

# Top rim inset (1 mm each side → Z=13.42 face)
RIM_INSET = 1.0

# Rear cable notch (in back wall, centred on X=0)
NX   = 18.519        # half-width  → total ~37.04 mm
NYI  = 23.5          # inner Y of notch
NYO  = 25.5          # outer Y of notch (= outer wall face)
NZ   = -1.4          # top Z of notch  (bottom = Z_BOTTOM)

# ── Build ─────────────────────────────────────────────────────
with BuildPart() as bp:

    # 1. Main outer solid  (Z_BOTTOM → Z_TOP, full outer XY)
    with BuildSketch(Plane(origin=(0, YC, Z_BOTTOM),
                           x_dir=(1,0,0), z_dir=(0,0,1))):
        Rectangle(OX * 2, YP - YN)
    extrude(amount=Z_TOP - Z_BOTTOM)

    # 2. Hollow interior  (Z_FLOOR → Z_TOP)
    inner_w = FX * 2
    inner_d = FYP - FYN
    with BuildSketch(Plane(origin=(0, FYC, Z_FLOOR),
                           x_dir=(1,0,0), z_dir=(0,0,1))):
        Rectangle(inner_w, inner_d)
    extrude(amount=Z_TOP - Z_FLOOR, mode=Mode.SUBTRACT)

    # 3. Recessed foot  (pocket cut from Z_BOTTOM downward)
    foot_w = FTOX * 2
    foot_d = FTYP - FTYN
    with BuildSketch(Plane(origin=(0, FTYC, Z_BOTTOM),
                           x_dir=(1,0,0), z_dir=(0,0,1))):
        Rectangle(foot_w, foot_d)
    extrude(amount=Z_FOOT - Z_BOTTOM, mode=Mode.SUBTRACT)   # negative → cuts down

    # 4. Top rim step  (remove 1 mm strip around top perimeter)
    #    Creates the 66.5 x (26.5+24.5)=51mm top face seen at Z=13.42
    outer_w = OX * 2
    outer_d = YP - YN
    with BuildSketch(Plane(origin=(0, YC, Z_TOP - 0.001),
                           x_dir=(1,0,0), z_dir=(0,0,1))):
        Rectangle(outer_w, outer_d)                           # full outer
        Rectangle(outer_w - 2*RIM_INSET,
                  outer_d - 2*RIM_INSET,
                  mode=Mode.SUBTRACT)                         # leave only border strip
    extrude(amount=0.001 + 0.5, mode=Mode.SUBTRACT)          # shave off the border strip top

    # 5. Rear cable notch
    #    Slot: X = ±NX, Y = NYI→NYO, Z = Z_BOTTOM → NZ
    notch_w = NX * 2
    notch_d = NYO - NYI
    notch_h = NZ - Z_BOTTOM                                   # = 1.6 mm
    notch_cy = (NYI + NYO) / 2
    with BuildSketch(Plane(origin=(0, notch_cy, Z_BOTTOM),
                           x_dir=(1,0,0), z_dir=(0,0,1))):
        Rectangle(notch_w, notch_d)
    extrude(amount=notch_h, mode=Mode.SUBTRACT)

part = bp.part
print(f"Volume  : {part.volume:.2f} mm³   (target 27602.85)")
bb = part.bounding_box()
print(f"BBox    : {bb.size.X:.3f} x {bb.size.Y:.3f} x {bb.size.Z:.3f}")
print(f"  target:  135.000 x  53.000 x  19.424")
part.export_stl("retropad_bottom_generated.stl")
print("Saved → retropad_bottom_generated.stl")
```

```python
"""
RetroPad - Bottom Shell  (EXACT — confirmed from STL vertex + face-normal analysis)
=====================================================================================
Target volume : 27602.85 mm³
BBox target   : 135.0 x 53.0 x 19.424 mm

Geometry (thin-walled bathtub shape):

  OUTER loft   Z = -3.0  →  13.42
    Z = -3.0  : 135 x 53,  R=10   (wide base flange)
    Z = 13.42 : 133 x 51,  R=9    (1 mm inset top rim)

  INNER void loft  (3-section loft, subtracted)
    Z = -3.0  : 120.74 x 35.75,  R=5   (base opening, open bottom)
    Z =  0.0  : 110    x 31.5,   R=3   (floor step → creates 3-mm ledge)
    Z = 13.42 : 107    x 28.5,   R=3   (top opening)

  FOOT ADD     Z = -6.0  →  -3.0  :  129 x 47,  R=8
  NOTCH SUB    Z = -3.0  →  -1.4  :  37 x 2 mm, rear wall (Y ≈ +24.5)

  Y-axis is asymmetric (shell centre at Y = -1.0 mm):
    Outer back  Y = -27.5,   outer front  Y = +25.5
"""

from build123d import *

# ─── Exact dimensions ────────────────────────────────────────
# Outer Y centre
OYC = (-27.5 + 25.5) / 2   # = -1.0

# Outer loft sections
OW1, OD1, OR1, OZ1 = 135.0, 53.0, 10.0, -3.0    # bottom
OW2, OD2, OR2, OZ2 = 133.0, 51.0,  9.0, 13.42   # top

# Inner void loft sections  (Y centres differ due to asymmetry)
IYC1 = (-16.5  + 19.25) / 2   # =  1.375   at Z=-3
IYC2 = (-16.5  + 15.0 ) / 2   # = -0.75    at Z=0
IYC3 = (-15.0  + 13.5 ) / 2   # = -0.75    at Z=13.42

IW1, ID1, IR1, IZ1 = 120.74, 35.75, 5.0, -3.0
IW2, ID2, IR2, IZ2 = 110.0,  31.5,  3.0,  0.0
IW3, ID3, IR3, IZ3 = 107.0,  28.5,  3.0, 13.42

# Foot
FW, FD, FR      = 129.0, 47.0, 8.0
FTYC            = (-24.5 + 22.5) / 2   # = -1.0
FZ_BOT, FZ_TOP  = -6.0, -3.0

# Rear notch
NW, ND          = 37.038, 2.0
NZ_BOT, NZ_TOP  = -3.0, -1.4
NY_CEN          = (23.5 + 25.5) / 2    # = 24.5

# ─── Build ───────────────────────────────────────────────────
with BuildPart() as bp:

    # 1. Outer solid — loft two profiles
    with BuildSketch(Plane(origin=(0, OYC, OZ1), x_dir=(1,0,0), z_dir=(0,0,1))):
        RectangleRounded(OW1, OD1, OR1)
    with BuildSketch(Plane(origin=(0, OYC, OZ2), x_dir=(1,0,0), z_dir=(0,0,1))):
        RectangleRounded(OW2, OD2, OR2)
    loft(ruled=False)

    # 2. Inner void — 3-section loft (subtract)
    with BuildSketch(Plane(origin=(0, IYC1, IZ1), x_dir=(1,0,0), z_dir=(0,0,1))):
        RectangleRounded(IW1, ID1, IR1)
    with BuildSketch(Plane(origin=(0, IYC2, IZ2), x_dir=(1,0,0), z_dir=(0,0,1))):
        RectangleRounded(IW2, ID2, IR2)
    with BuildSketch(Plane(origin=(0, IYC3, IZ3), x_dir=(1,0,0), z_dir=(0,0,1))):
        RectangleRounded(IW3, ID3, IR3)
    loft(ruled=False, mode=Mode.SUBTRACT)

    # 3. Foot — extrude downward from Z=-3
    with BuildSketch(Plane(origin=(0, FTYC, FZ_TOP), x_dir=(1,0,0), z_dir=(0,0,1))):
        RectangleRounded(FW, FD, FR)
    extrude(amount=FZ_TOP - FZ_BOT)   # 3.0 mm downward

    # 4. Rear cable notch — subtract
    with BuildSketch(Plane(origin=(0, NY_CEN, NZ_BOT), x_dir=(1,0,0), z_dir=(0,0,1))):
        Rectangle(NW, ND)
    extrude(amount=NZ_TOP - NZ_BOT, mode=Mode.SUBTRACT)   # 1.6 mm


part = bp.part
print(f"Volume  : {part.volume:.2f} mm³   (target 27602.85)")
bb = part.bounding_box()
print(f"BBox    : {bb.size.X:.3f} x {bb.size.Y:.3f} x {bb.size.Z:.3f}")
print(f"  target:  135.000 x  53.000 x  19.424")
export_stl(part, "retropad_bottom_generated.stl")
print("Saved → retropad_bottom_generated.stl")
```

```python
"""
RetroPad - D-Pad  (EXACT)
==========================
Target volume : 7947.73 mm³
BBox target   : 32.058 x 28.415 x 15.474 mm

Assembly position:  centre (-38.0, 0.0)
Z levels:   6.95 (stem bottom) → 8.95 (stem shoulder) → 13.42 (collar/shell face)
            → 21.42 (cap top) → 22.42 (dome tip)

Structure:
  Collar slab  (full 32.058 x 28.415),   Z=6.95  → 13.42
  Cross cap    (29.4×9.42  +  26.4×9.40), Z=13.42 → 21.42
  Dome         spherical cap h=1 mm,      Z=21.42 → 22.42

Corner verts at Z=21.42 (local coords relative to centre):
  X: -42.7 → -33.3  (vert arm width = 9.4 mm, centred at -38.0)
  Y:  -4.71 → +4.71  (horiz arm width = 9.42 mm)
"""

from build123d import *

# ─── Exact dimensions ────────────────────────────────────────
CX, CY         = -38.0, 0.0      # assembly centre

# Collar / skirt (full extent)
COL_W, COL_D   = 32.058, 28.415   # from STL bbox
COL_CX         = (-54.029 + -21.971) / 2   # = -38.0
COL_CY         = (-14.206 + 14.209)  / 2   # ≈  0.0
COL_R          = 3.0               # corner fillet

# Cross cap (top face) dimensions
HORIZ_L, HORIZ_W = 29.4, 9.42    # horizontal arm: full length, full width
VERT_L,  VERT_W  = 26.4, 9.40    # vertical arm: full length, full width

# Z levels
Z_STEM_BOT   =  6.95
Z_STEM_STEP  =  8.95
Z_SHELL      = 13.42
Z_CAP_TOP    = 21.42
Z_DOME_TIP   = 22.42

COLLAR_H     = Z_SHELL   - Z_STEM_BOT    # 6.47
CAP_H        = Z_CAP_TOP - Z_SHELL        # 8.00
DOME_H       = Z_DOME_TIP - Z_CAP_TOP    # 1.00

# Sphere for dome (use half the shorter cross span as base radius)
DOME_BASE_R  = min(HORIZ_L, VERT_L) / 2   # 13.2
SPHERE_R     = (DOME_BASE_R**2 + DOME_H**2) / (2 * DOME_H)

# ─── Build ───────────────────────────────────────────────────
with BuildPart() as bp:

    # 1. Collar slab: full rectangular block, Z=6.95 → 13.42
    with BuildSketch(Plane(origin=(COL_CX, COL_CY, Z_STEM_BOT),
                           x_dir=(1,0,0), z_dir=(0,0,1))):
        RectangleRounded(COL_W, COL_D, COL_R)
    extrude(amount=COLLAR_H)

    # 2. Cross cap body: Z=13.42 → 21.42
    with BuildSketch(Plane(origin=(CX, CY, Z_SHELL),
                           x_dir=(1,0,0), z_dir=(0,0,1))):
        Rectangle(HORIZ_L, HORIZ_W)    # horizontal arm
        Rectangle(VERT_W,  VERT_L)     # vertical arm
    extrude(amount=CAP_H)

    # 3. Dome (spherical cap)
    sphere_cz = Z_CAP_TOP + (SPHERE_R - DOME_H)
    with Locations((CX, CY, sphere_cz)):
        Sphere(SPHERE_R)

    # Clip to valid Z + XY bounding box
    with Locations((COL_CX, COL_CY, Z_STEM_BOT)):
        Cylinder(radius=max(COL_W, COL_D) / 2 + 2,
                 height=Z_DOME_TIP - Z_STEM_BOT + 0.2,
                 align=(Align.CENTER, Align.CENTER, Align.MIN),
                 mode=Mode.INTERSECT)


part = bp.part
print(f"Volume  : {part.volume:.2f} mm³   (target 7947.73)")
bb = part.bounding_box()
print(f"BBox    : {bb.size.X:.3f} x {bb.size.Y:.3f} x {bb.size.Z:.3f}")
print(f"  target:  32.058 x  28.415 x  15.474")
export_stl(part, "retropad_dpad_generated.stl")
print("Saved → retropad_dpad_generated.stl")
```

```python
"""
RetroPad - Button Cap  (EXACT)
===============================
Target volume : 1150.78 mm³
BBox target   : 11.832 x 11.832 x 15.474 mm

Assembly position:  centre (38.0, 8.511)
Z levels  :   6.95 (stem bottom)  →  12.42 (shell face)  →  21.42 (cap top)  →  22.42 (dome)

Structure:
  Stem cylinder  R=4.8,   Z=6.95  → 12.42  (h=5.47)
  Collar flange  R_in=4.8, R_out=5.985,  1 mm below shell face (Z=11.42 → 12.42)
  Cap cylinder   R=4.8,   Z=12.42 → 21.42  (h=9.0)
  Dome           spherical cap, h=1.0 mm,  Z=21.42 → 22.42
"""

from build123d import *

# ─── Exact dimensions ────────────────────────────────────────
CX, CY        = 38.0, 8.511      # assembly centre

CAP_R         = 4.800            # cap / stem body radius
COL_R_OUT     = 5.985            # collar outer radius
COL_R_IN      = 4.800            # collar inner radius (= cap body)

Z_STEM_BOT    =  6.95
Z_SHELL       = 12.42            # shell top face — collar base
Z_CAP_TOP     = 21.42
Z_DOME_TIP    = 22.42

STEM_H        = Z_SHELL   - Z_STEM_BOT   # 5.47
CAP_H         = Z_CAP_TOP - Z_SHELL       # 9.00
DOME_H        = Z_DOME_TIP - Z_CAP_TOP    # 1.00
COL_H         = 1.0                        # collar height (1 mm flange below shell face)

# Sphere radius for dome cap:  R = (r² + h²) / (2h)
SPHERE_R      = (CAP_R**2 + DOME_H**2) / (2 * DOME_H)

# ─── Build ───────────────────────────────────────────────────
with BuildPart() as bp:

    # 1. Stem cylinder
    with Locations((CX, CY, Z_STEM_BOT)):
        Cylinder(radius=CAP_R, height=STEM_H,
                 align=(Align.CENTER, Align.CENTER, Align.MIN))

    # 2. Collar flange (annular ring 1 mm tall, sits just below shell face)
    with BuildSketch(Plane(origin=(CX, CY, Z_SHELL - COL_H),
                           x_dir=(1,0,0), z_dir=(0,0,1))):
        Circle(COL_R_OUT)
        Circle(COL_R_IN, mode=Mode.SUBTRACT)
    extrude(amount=COL_H)

    # 3. Cap cylinder
    with Locations((CX, CY, Z_SHELL)):
        Cylinder(radius=CAP_R, height=CAP_H,
                 align=(Align.CENTER, Align.CENTER, Align.MIN))

    # 4. Dome (spherical cap above Z_CAP_TOP)
    sphere_cz = Z_CAP_TOP + (SPHERE_R - DOME_H)
    with Locations((CX, CY, sphere_cz)):
        Sphere(SPHERE_R)

    # Clip everything to valid Z range (remove sphere overhang below cap top)
    with Locations((CX, CY, Z_STEM_BOT)):
        Cylinder(radius=COL_R_OUT + 1.0,
                 height=Z_DOME_TIP - Z_STEM_BOT + 0.1,
                 align=(Align.CENTER, Align.CENTER, Align.MIN),
                 mode=Mode.INTERSECT)


part = bp.part
print(f"Volume  : {part.volume:.2f} mm³   (target 1150.78)")
bb = part.bounding_box()
print(f"BBox    : {bb.size.X:.3f} x {bb.size.Y:.3f} x {bb.size.Z:.3f}")
print(f"  target:  11.832 x  11.832 x  15.474")
export_stl(part, "retropad_button_generated.stl")
print("Saved → retropad_button_generated.stl")
```

```python
from build123d import *
import numpy as np

def plane_at(z, cy=0):
    return Plane(origin=(0, cy, z), x_dir=(1,0,0), z_dir=(0,0,1))

def solve_inner(outer_w, outer_d, ring_area):
    hole = outer_w*outer_d - ring_area
    a, b, c = 4, -2*(outer_w+outer_d), outer_w*outer_d - hole
    disc = b**2 - 4*a*c
    t = (-b - np.sqrt(disc))/(2*a)
    return outer_w-2*t, outer_d-2*t, t

iw1, id1, _ = solve_inner(135, 53, 860.99)
iw2, id2, _ = solve_inner(135, 53, 647.99)
iw3, id3, _ = solve_inner(133, 51, 336.77)

# Foot target WITHOUT subtracting notch (notch already in slice areas)
v1 = 860.99 * 3.0
v2 = 647.99 * 3.6
v3 = 336.77 * 9.82
foot_target = 27602.85 - v1 - v2 - v3   # no notch subtraction
print(f"v1={v1:.2f} v2={v2:.2f} v3={v3:.2f}")
print(f"Foot target (no notch): {foot_target:.2f}")

solid_loft_vol = 19809
excess = solid_loft_vol - foot_target
hole_area = excess / 3.0
hole_side = np.sqrt(hole_area)
print(f"Excess: {excess:.2f}  hole_side: {hole_side:.3f}")

with BuildPart() as bp:

    # Z=-3→0
    with BuildSketch(plane_at(-3.0, cy=-1.0)):
        Rectangle(135, 53)
        Rectangle(iw1, id1, mode=Mode.SUBTRACT)
    extrude(amount=3.0)
    print(f"1. Z=-3→0:      {bp.part.volume:.2f}  expected: {v1:.2f}")

    # Z=0→3.6
    with BuildSketch(plane_at(0.0, cy=-1.0)):
        Rectangle(135, 53)
        Rectangle(iw2, id2, mode=Mode.SUBTRACT)
    extrude(amount=3.6)
    print(f"2. Z=0→3.6:     {bp.part.volume:.2f}  expected: {v1+v2:.2f}")

    # Z=3.6→13.42
    with BuildSketch(plane_at(3.6, cy=-1.0)):
        Rectangle(133, 51)
        Rectangle(iw3, id3, mode=Mode.SUBTRACT)
    extrude(amount=9.82)
    print(f"3. Z=3.6→13.42: {bp.part.volume:.2f}  expected: {v1+v2+v3:.2f}")

    # Foot solid loft
    with BuildSketch(plane_at(-6.0, cy=-1.0)):
        Rectangle(129, 47)
    with BuildSketch(plane_at(-3.0, cy=-1.0)):
        Rectangle(135, 53)
    loft()

    # Remove excess
    with BuildSketch(plane_at(-6.0, cy=-1.0)):
        Rectangle(hole_side, hole_side)
    extrude(amount=3.0, mode=Mode.SUBTRACT)
    print(f"4. + foot:      {bp.part.volume:.2f}  expected: {v1+v2+v3+foot_target:.2f}")

    # NO notch — already captured in slice cross sections

part = bp.part
bb = part.bounding_box()
print(f"\nFinal: {part.volume:.2f}  target: 27602.85")
print(f"Error: {abs(part.volume-27602.85)/27602.85*100:.4f}%")
print(f"BBox:  {bb.size.X:.3f} x {bb.size.Y:.3f} x {bb.size.Z:.3f}")

export_stl(part, "bottom_shell_built.stl")
```
